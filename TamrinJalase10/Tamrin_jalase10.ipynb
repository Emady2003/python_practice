```python
def findM(file1,file2):
    word_list = []
    word_list2 = []
    lines = file1.readlines()
    for line in lines:
        word_list.extend(line.split())
    lines = file2.readlines()
    for line in lines:
        word_list2.extend(line.split())

    intersection = set(word_list) & set(word_list2)
    return(list(intersection))
    

findM(open("001.txt","r"),open("002.txt","r"))
```

```python
def copy_file(file1,file2):
    list1 = []
    for line in file1:
        list1.append(line.strip("\n"))
    file2.write("")
    len_list = int(len(list1))
    while len_list > 0 :
        file2.seek(0,0)
        file2.write(f"\n{list1[-len_list]}")
        len_list -= 1
    return(list1,)

    

copy_file(open("001.txt","r+"),open("002.txt","w"))
```
